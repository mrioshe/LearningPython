### <font color='blue'>**6. Velocidad de caída de un paracaidista** </font> <br>
La velocidad de un paracaidista de masa m que parte del reposo, el cual
cae con un coeficiente de arrastre C, se determina según la siguiente
relación:

$$ V(t)=\dfrac{mg}{C}\left[1-e^{-\left( \frac{C}{m}\right) t} \right]  $$

Una aproximación numérica a esta solución se puede obtener mediante la
siguiente expresión:

$$ V_a\left(t+\Delta t \right) =V_a\left( t\right) +\left( g-\dfrac{C\ V_a\left( t\right)}{m}\right)  \Delta t $$

Diseñe un algoritmo que determine la velocidad mediante la ecuación 2
para un tiempo final y un paso temporal determinados, así como los
errores absoluto y relativo de la aproximación. Use una masa de 68.1 kg, y un coeficiente de arrastre de 12.5 Ns/m, así como una aceleración gravitacional de 9.81 m/s². Recuerde que la velocidad incial vertical de caída es cero. <br> <br>
<font color='green'>Resuelva el problema usando la estructura **para**, y vectores (arreglos unidimensionales) para representar el tiempo de caída y la velocidad aproximada.

```python
# Importamos librerías:
import numpy as np

# inicialización de variables:
g = 9.81 # m/s^2

# Entrada de datos:

m = float(input('Ingrese la masa de paracaídista [Kg]:   '))
c = float(input('Ingrese coeficiente de arrastre [Ns/m]:   '))
t_c = float(input('Ingrese el tiempo de caída [s]: '))
Dt= float(input('Ingrese paso temporal [s]:   '))

# Creación del vector:

t = np.arange(0, t_c + Dt, Dt)
v_a = np.zeros(len(t))

# Procesos:

v = (m * g / c)*(1-np.exp(-c*t_c/m)) # Velocidad real

for i in range(1, len(v_a)):

    v_a[i] = v_a[i-1]  + (g-(c*v_a[i-1]/m))*Dt 
    
error_abs = v - v_a[-1]
error_relativo = 100 * abs(error_abs/v)

# Salidas:

print('\nLa velocidad a los %d s es: %.2f ' % (t_c, v_a[-1]))
print('\nEl error absoluto es: %.2f' % error_abs)
print('El error relativo es: %.2f' % error_relativo)

    
```

### <font color='blue'>**8. Clasificación de estudiantes según estatura** </font> <br>

Calcular la estatura media de los alumnos de una clase, y determinar
cuántos están por encima (altos) y cuántos por debajo de ésta (bajos). <br>
Realizar la prueba de escritorio con el siguiente vector de estaturas:<br>
estaturas = \[ 1.60   1.56   1.75   1.58   1.52   1.82   1.67   1.74   1.69   1.71 \]
<br> <br>
<font color='green'>Resuelva el problema de dos formas: <br>
* Use un vector.
* Use listas y programación compacta.

```python
# Versión con un vector:
# importamos la librería:
import numpy as np

# Inicilización de variables:
altos, bajos = 0,0

#Entrada de datos:
estaturas = np.array(eval('[' + input('Ingrese las estaturas de los alumnos, separadas por comas:  ') + ']'))

#procesos y salidas:

promedio = np.mean(estaturas )

for i in range (len(estaturas)):
    
    if estaturas[i] >= promedio:
        altos += 1
    else: 
        bajos += 1

print('\nEl número de estudiantes que están igual o por encima de la media son: ', altos)
print('El número de estudiantes que están por debajo de la media son: ', bajos)
print('La media es: %.2f ' %promedio)
```

### <font color='blue'>**9. Clasificación de los elementos de una matriz** </font> <br>

Dada una matriz M, de dimensión (m,n), desarrollar un algoritmo que
identifique la cantidad de valores positivos y negativos en ésta, y además
calcule la suma de todos los valores positivos, y la de todos los valores
negativos.<br>
Realice la prueba con la siguiente matriz:
$$\texttt{M} =  \left[ 
				\begin{array}{cccc}
				3	&	-2	&	8	&	-9	\\
				5	&	14	&	-0.5	&	1	\\
				-4	&	7	&	22.3	&	6.7	\\
				\end{array} \right] $$
<br> <br>
<font color='green'>Resuelva el problema de dos formas: <br>
* Use una matriz.
* Use listas y programación compacta.

```python
#Versión con matrices:
# Importamos librerías:
import numpy as np

# Entrada de datos:

datos = np.array(eval('[' + input('Ingrese la matriz como listas de una de lista separadas por coma:    ') + ']'))

# Inicialización de variables:

contador_positivos,contador_negativos,suma_positivos, suma_negativos = 0,0,0,0

# Procesos:

m,n = np.shape(datos)[0], np.shape(datos)[1]

for i in range (m):
    for j in range (n):
        if datos[i,j] >= 0:
            contador_positivos += 1
            suma_positivos = suma_positivos + datos[i,j]
        else:
            contador_negativos += 1
            suma_negativos = suma_negativos + datos[i,j]

print('\nEl total de números positivos en la matriz es %.1f y su suma es igual a %.1f' %(contador_positivos, suma_positivos))
print('El total de números negativosen la matriz es %1.f y su suma es igual a %.1f' %(contador_negativos, suma_negativos))

```

### <font color='blue'>**10. Gestión de reservas en una aerolínea** </font> <br>

Una aerolínea gestiona su sistema de reservas mediante un arreglo de tres
dimensiones, en el cual las filas corresponden a la fila dentro del avión, las
columnas al número del asiento, y cada matriz corresponde a un número
de vuelo. <br>
Cada elemento del arreglo tiene un valor de 1 (asiento disponible) ó 0
(asiento ocupado). <br>
El usuario ingresa el vuelo, la fila y el número de asiento. <br>
Diseñe un algoritmo que informe a los clientes sobre la disponibilidad de
un asiento solicitado, y actualice la base de datos según la selección de
asiento. Trabaje con el arreglo tridimensional dado por el siguiente código: 5 vuelos, cada uno con 25 filas y 8 columnas. <br>

`reservas_r = np.random.random((5, 25, 8))`<br>
`reservas = np.round(reservas_r)` <br>

El programa debe comenzar por presentar la siguiente información al usuario:<br>
***
SISTEMA DE RESERVAS

Los vuelos disponibles son: VA01, VA02, VA03, VA04, y VA05

Cada vuelo tiene 25 filas

En cada fila se tienen asientos enumerados desde el A hasta el H
***
El vuelo VA03, fila 14, asiento F corresponde a la fila 13, columna 5, del vuelo 2: reservas[2, 13, 5]<br>

El programa debe informar sobre los asientos disponibles en cada vuelo, después de que se hace una reserva.

```python
#Creación del arreglo de reservas
reservas_r = np.random.random((5, 25, 8))
reservas = np.round(reservas_r)

print('_'* 70)
print(''' SISTEMA DE RESERVAS

Los vuelos disponibles son: VA01, VA02, VA03, VA04, y VA05

Cada vuelo tiene 25 filas

En cada fila se tienen asientos enumerados desde el A hasta el H ''')
print('_'* 70)

# Diccionarios par identificar la reserva:
vuelos ={'VA01':0,'VA02':1,'VA03':2,'VA04':3,'VA05':4 }
columnas={'A': 0,'B': 1,'C': 2,'D': 3,'E': 4,'F': 5,'G': 6,'H': 7}

# Entrada de datos
vuelo = input('\nNúmero de vuelvo:  ')

# Gestión de la reserva:

while True:
    fila = int(input('Número de filas:  ')) - 1
    col = input('Número de asiento:   ')
    if reservas[vuelos[vuelo], fila, columnas [col]] == 1 :
        print('\nEL asiento seleccionado está disponible. ')
        print('\nLa reserva se realizó correctamente. ')
        reservas[vuelos[vuelo], fila, columnas [col]] = 0
        break
    else: 
        print('\nEL asiento seleccionado no está disponible ')
        it = input('\n¿Desea seleccionar otro asiento? S/N:     ' ).lower()
        if it == 's':
            continue
        else: 
            break     
# Informe de asientos disponibles en cada vuelo:

l_vuelos = list (vuelos.keys())
print('_'* 70)
print('\nCantidad de asientos disponibles \n')
print('vuelo', 'No. de asientos', sep='\t')
for i in range (len(vuelos)):
    print(l_vuelos[i],np.sum(reservas[i,:,:]), sep='\t')
```
